```python
import json

f = open('all_user.json')
data = json.load(f)
    
user_data = pd.DataFrame(data=data)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x=user_data['sales'],y=user_data['n_markets'],)
```

```python
user_data.sort_values(by='sales',ascending=False)
```

```python
user_data['markets'].explode().value_counts()[:10].plot(title='Most Popular Markets',xlabel='Dark Web Marketplace',ylabel='Frequency(No. of Vendors)',linewidth=2,figsize=(10,5),kind='bar')
```

```python
user_data['n_markets'].value_counts()[1:11].plot(title='Vendor Frequency Distribution: Number of Markets',xlabel='No. of markets vendor is active in',ylabel='Frequency (No. of Vendors)',linewidth=2,figsize=(10,5),kind='bar')
```

```python
user_data['ships_from'].value_counts()[1:11].plot(title='Countries with Highest Number of Vendors',xlabel='Country of Origin',ylabel='Frequency (No. of Vendors)',linewidth=2,figsize=(10,5),kind='bar')
```

```python
user_data['products'].value_counts()[1:10].plot(title='Most Popular Products to Sell',xlabel='Product Type',ylabel='Frequency (No. of Vendors)',linewidth=2,figsize=(10,5),kind='bar')
```

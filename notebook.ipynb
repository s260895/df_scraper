```python
import pandas as pd
import numpy as np
```

```python
user_data = pd.read_csv('data/all_users.csv')
del user_data['Unnamed: 0']
```

```python
user_data['n_markets'].value_counts()[1:11].plot(title='Frequency Distribution: Number of Markets',xlabel='No. of markets that vendor is active in',ylabel='Frequency (No. of Vendors)',linewidth=2,figsize=(10,5),kind='bar')
```

```python
user_data['ships_from'].value_counts()[1:11].plot(title='Countries with Highest Number of Vendors',xlabel='Country of Origin',ylabel='Frequency (No. of Vendors)',linewidth=2,figsize=(10,5),kind='bar')
```

```python
user_data['products'].value_counts()[1:11].plot(title='Most Popular Products to Sell',xlabel='Product Type',ylabel='Frequency (No. of Vendors)',linewidth=2,figsize=(10,5),kind='bar')
```

```python
from tbselenium.tbdriver import TorBrowserDriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import os
import time
```

```python
USERNAME = "jheronimus"
PASSWORD = "dataforensics"
PASSCODE = "8899"

# Settings
PAGES_NUMBER = 1

# URLS
ALTERNATE_URLS = ['http://silkroad4ow7pliq.onion',
                 "http://silkroad4uqi7hey.onion"]

URL = ALTERNATE_URLS[1]
URL_LOGIN = URL + "/?login"
# URL_CAPTCHA = URL + "challenge.php"


# #log in and solve captcha

def visit_and_login(driver):
    
    
    #driver.get(URL_CAPTCHA)
    
    
    driver.get(URL_LOGIN)

    # Log in and solve the captcha
    #WebDriverWait(driver, 120).until(EC.title_contains("login"))
    user_field = driver.find_element_by_name("username")
    pass_field = driver.find_element_by_name("password")
    login_button = driver.find_element_by_name('login')
    
    user_field.send_keys(USERNAME)
    time.sleep(4)
    pass_field.send_keys(PASSWORD)
    time.sleep(4)
    
    login_button.click()
    
    WebDriverWait(driver, 20)
    
    
    return driver

#go trough pages and store information
def get_pages(driver):
    print("Starting the crawling")
    for i in range(1, PAGES_NUMBER + 20):
        site_page = URL + '/?road=&reqpage=' + str(i)
        driver.get(site_page)
        time.sleep(2)

        filename = 'silkroad/'+ 'page_'+ str(i) + '.html'
        with open("/home/levi/" + filename, "w") as f:
            f.write(driver.page_source)
            print('dowloaded' ,i, 'page')
```

```python
with TorBrowserDriver("/home/levi/tor-browser_en-US/") as driver:
    driver = visit_and_login(driver)
    get_pages(driver)
```

```python
import os
from bs4 import BeautifulSoup as bs

#list that will contain urls
vendor_urls = []


#set directory containing html pages
DIR = '/home/levi/silkroad'

#for each html page
for page in os.listdir(DIR):
    
    html_file = os.path.join(DIR, page)#link to html
    
    #open file with beautiful soup
    with open(html_file) as f:
        soup = bs(f, 'html.parser')
        
        vendors = soup.find_all('span', id='ah_name') #find lines containing vendor links
        
        vendors = list(vendors) #make it a list so we can iterate
        
        for vendor in vendors:
            url = vendor.a['href'] #extract only the hyperlink
            vendor_urls.append(url)
```

```python
vendor_urls = ['http://silkroad4uqi7hey.onion' +'/'+ i for i in vendor_urls]
```

```python
#function to download vendor pages
def get_vendor_pages(driver, vendor_urls):
    print("Starting the crawling")
    for i in vendor_urls:
        site_page = i
        driver.get(site_page)
        time.sleep(3)

        filename = 'silkroad/'+ str(i[40:]) + '.html'
        with open("/home/levi/" + filename, "w") as f:
            f.write(driver.page_source)
            print('dowloaded' ,i[40:])

            
with TorBrowserDriver("/home/levi/tor-browser_en-US/") as driver:
    driver = visit_and_login(driver)
    get_vendor_pages(driver, vendor_urls)
```
